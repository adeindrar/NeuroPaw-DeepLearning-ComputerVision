# Import Library

> import library yang diperlukan

```python
# Import libraries

import numpy as np
import json
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model

# Untuk menghilangkan warning pada notebook
import warnings
warnings.filterwarnings("ignore")
```

# Load Models

> Mari kita load model dan clas yang telah disimpan sebelumnya

```python
# Load files!

with open('class_name.txt', 'r') as file_1:
    class_name = json.load(file_1)

model_ann = tf.keras.models.load_model('saved_model')
```

model dan class telah berhasil di load!

# Inferencing

Mari kita buat function untuk memprediksi gambar baru dari website

```python
# Create a function for test a new image

def prediction(file):
  img = tf.keras.utils.load_img(file, target_size=(200,200))
  x = tf.keras.utils.img_to_array(img)/255

  plt.imshow(img)

  x = np.expand_dims(x, axis=0)

  images = np.vstack([x])
  y_pred_proba = model_ann.predict(images)
  y_pred_class = np.argmax(y_pred_proba[0])

  y_pred_class_name = class_name[y_pred_class]

  print('Prediction : {}'.format(y_pred_class_name))
```

Kita prediksi gambar pertama, gambar ini didownload dari google dengan mengambil link gambar dan mendownloadnya
